#**Jeovany Dominique**#
##*and my sad Knicks fandom*#

<p> I have been a huge knicks fan for quite sometime now, ever since I was little in fact </p>

<p> One thing that has been a constant in that time is that the New York Knickbockers have not been very good at the game of basketball. </p>

<p> As of late though they have turned themselves around and become a respectable and competitive franchise which is reassuring for the future no doubt.</p>

<p> I say all of that to call to attention the fact that I have been to 13 knicks games in the last 5 years, and each time I go to Madison Square Garden, I find myself on the receiving end of a loss</p>

<p> I feel this team lives to break my heart</p>
